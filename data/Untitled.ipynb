```python
import pandas as pd
```

```python
df = pd.read_csv("breweries.csv")
```

```python
# Remove duplicate rows
df.drop_duplicates
```

```python
df = df[df['longitude'].notna()]
df
```

```python
df = df[df["country"]== "United States"]
```

```python
df = df.fillna("Not Available")
```

```python
df.head()
```

```python
df.head()
```

```python
df["region"] = df["state"]
```

```python
df.head()
```

```python
df["region"].str.replace("Indiana","Midwest")
```

```python
df["region"] = df["region"].str.replace("Indiana","Midwest")
```

```python
regions = pd.read_csv("regions.csv")
```

```python
regions.head()
```

```python
regions["state"] = regions["State"]
```

```python
final_data = pd.merge(df, regions, how="left", on=["state"])
```

```python
final_data.head()
```

```python
address = final_data["street"] + " " + final_data["city"] + " " + final_data["state"] + " " + final_data["postal_code"]
address
```

```python
final_data['address'] = address
final_data
```

```python
final_data = final_data[['name', 'brewery_type', 'address', 'state', 'phone', 'website_url', 'country', 'Region', 'Division', 'longitude', 'latitude']]
final_data
```

```python
final_data = final_data.rename(columns={'Region': 'region', 'Division': 'division'})
final_data
```

```python
final_data = final_data.dropna(how='any')
final_data
```

```python
final_data.to_csv("breweries_clean.csv",index = False)
```
